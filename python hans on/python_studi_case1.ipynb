```python
df_1 = pd.read_excel("/Users/newuser/ali/project/binar/notebook/dataset/World Bank Education.xls",sheet_name='Data')
```

## change name coloumn use row 3 / index 2

```python
df_1.columns = df_1.loc[2]
```

## reset index and setup row starting number row 3

```python
df_world = df_1.loc[3:]
df_world.reset_index(drop=True)
```

## filter country indenesia

```python
df_filter = df_world.loc[df_world['Country Code'] == 'IDN'] 
```

## create change type to str biar bisa di rename lower case atau lower case spesifik coloumn

```python
df_filter.columns = df_filter.columns.astype(str)
```

```python
#lower case coloumn
df_filter.columns = map(str.lower, df_filter.columns)
df_filter.columns = df_filter.columns.str.replace(' ', '_')
```
